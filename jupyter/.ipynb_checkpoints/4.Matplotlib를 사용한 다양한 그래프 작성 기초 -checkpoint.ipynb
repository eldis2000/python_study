# 4.Matplotlib를 사용한 다양한 그래프 작성 기초 

## 그래프 작성 / 불러오기 

```python
import matplotlib.pyplot as plt
```

```python
afig = plt.figure()

ax = fig.add_subplot(111) #행번호, 열번호, 그림번호 

data = [0,1]

ax.plot(data)

plt.show()
```

## 새 종이를 여러 등분으로 나누는 방법  

```python
fig = plt.figure()

x1 = fig.add_subplot(221) #2행 2열 1번째 그래프

x2 = fig.add_subplot(222) #2행 2열 2번째 그래프

x3 = fig.add_subplot(223) #2행 2열 3번째 그래프

plt.show()
```

## 그래프 스타일 변경 방법 

```python
plt.style.use('ggplot')

fig = plt.figure()

x = fig.add_subplot(111)

data = [0,1]

x.plot(data)

plt.show()
```

## 꺽은선 그래프 그리기 

```python
fig = plt.figure()

ax = fig.add_subplot(111)

x = [0,2,4]
y = [0,4,2]

ax.plot(x, y)

plt.show()
```

## 산포도 그래프 그리기 

```python
import numpy as np
```

```python
plt.style.use('ggplot')

np.random.seed(2)

x = np.arange(1,101)

y = 2 * x * np.random.rand(100)

fig = plt.figure()

ax = fig.add_subplot(111)

ax.scatter(x,y)

plt.show()
```

## 막대 그래프 그리기 

```python
plt.style.use('ggplot')

fig = plt.figure()

ax = fig.add_subplot(111)

x = [3,4]
y = [1,2]

ax.bar(x,y)
plt.show()
```

## 그래프 레이블 생성하기 

```python
plt.style.use('ggplot')

fig = plt.figure()

ax = fig.add_subplot(111)

labels = ['english', 'math']

x = [3,4]
y = [1,2]

ax.bar(x,y, tick_label = labels)
plt.show()
```

## 가로로 누운 그래프 그리기 

```python
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
labels = ['english','math']
x = [3,4]
y = [1,2]
ax.barh(x,y, tick_label = labels)
plt.show()
```

## 여러 데이터를 묶어 그리기 

```python
x = [1,2]
y1, y2, y3 = [1,2],[2,5],[3,7]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2
ax.bar(x,y1, width=w, label = 'english')
ax.bar(np.array(x) + w, y2, width = w, label = 'math')
ax.bar(np.array(x) + w *2, y3, width=w, label = 'korean')
ax.legend()
plt.show()
```

## 히스토그램 그리기 

```python
plt.style.use('ggplot')
data1 = 100
data2 = 10
np.random.seed(0)
x = np.random.normal(data1, data2, 1000)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(x)
plt.show()
```

## Boxplot 차트 그리기 

```python
plt.style.use('ggplot')
x = [1,2,3,3,11,20]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(x)
plt.show()
```

## Pie 차트 그리기 

```python
plt.style.use('ggplot')
labels = ['english', 'math', 'korean']
sizes = [45,30,25]
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
ax.pie(sizes, labels=labels)
plt.show()
```
